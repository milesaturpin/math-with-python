```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = np.arange(0, 10, 0.2)
data2 = np.sin(data1)
data3 = np.arange(0, 10, 0.2)
data4 = np.cos(data3)
#fig = plt.figure()
#ax = fig.add_subplot(111)
#ax.plot(x, y)
#plt.show()

def my_plotter(ax, data1, data2, param_dict):
    """ 
A helper function to make a graph.
    Parameters
    ----------
    ax : Axes
        The axes to draw to
    data1 : array
       The x data
    data2 : array
       The y data
    param_dict : dict
       Dictionary of kwargs to pass to ax.plot
    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,4))
my_plotter(ax1, data1, data2, {'marker':'x'})
my_plotter(ax2, data3, data4, {'marker':'o', 'linewidth':'2'})
```
